```python
# customer loyalty Convert tenure months to years for easier reading
df['Customer_Loyalty'] = df['Tenure_Months'] / 12
```

```python
df
```

```python
# Trused users which are below 3 support tickets
df['High_Support_Users'] = np.where(df['Support_Tickets'] > 3, 1, 0)
```

```python
df
```

```python
print(df[['Tenure_Months', 'Customer_Loyalty', 'Support_Tickets', 'High_Support_Users']].head())
```

```python
# Chart 1: Churn by Contract (Bar Chart)
sns.countplot(x='Contract', hue='Is_Churned', data=df)
plt.title('Churn by Contract Type')
plt.xlabel('Contract')
plt.ylabel('Count of Customers')
```

```python
# Chart 2: Monthly Charges vs Churn (Boxplot)
sns.boxplot(x='Is_Churned', y='Monthly_Charges', data=df)
plt.title('Distribution: Monthly Charges vs Churn')
```

```python
# Chart 3: Correlation Matrix (Heatmap)
numeric_df = df.select_dtypes(include=['float64', 'int64', 'int32'])
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Matrix')
```

```python
# Chart 4: Churn by Tenure (Line Chart)
tenure_churn = df.groupby('Tenure_Months')['Is_Churned'].mean()
sns.lineplot(x=tenure_churn.index, y=tenure_churn.values, color='red')
plt.title('Churn Rate by Tenure (Months)')
plt.ylabel('Churn Probability')
```
